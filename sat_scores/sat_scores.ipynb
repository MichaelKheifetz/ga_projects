One of the main assumptions in a dataset is Normality (whether the data follows a normal distribution). This can be tested using various Statistical tests (such as Kolmogorov-Smirnov, Anderson-Darling, etc tests), as well as looking at the histograms visually and getting a feel for their distribution. Additionally, the skewness and kurtosis of the distribution should be investigated.

##### 18. Plot some scatterplots. **BONUS**: Use a PyPlot `figure` to present multiple plots at once.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


data = pd.read_csv('sat_scores.csv')
data.columns = ["State", "Rate", "Verbal", "Math"]



fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(data.Rate, data.Math)
ax.set_title("Rate vs Math Scores")
ax.set_xlabel("Rate")
ax.set_ylabel("Math")
ax.set_xlim(0,100)

fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(data.Rate, data.Verbal)
ax.set_title("Rate vs Verbal Scores")
ax.set_xlabel("Rate")
ax.set_ylabel("Verbal")
ax.set_xlim(0,100)

fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(data.Math, data.Verbal)
ax.set_title("Math vs Verbal Scores")
ax.set_xlabel("Math")
ax.set_ylabel("Verbal")
ax.set_xlim(0,800)
```

##### 19. Are there any interesting relationships to note?

The Math and Verbal scores are positively correlated. This result has some intuition because students that are good in one subject (conditional) tend to on average be better than the average student in other subjects. For the Rate vs Math/Verbal relationships, there is a negative relationship. This might be more challenging to understand. One potential reason is there is probably some sort of initial data collection errors, it seems quiet unlikely that some states have participation Rates as low as 4% whereas others as high as 82% (this needs to be further investigated). There is also a potential selection bias that only stronger students in states where the participation Rates are low actually sit to take the exam and hence have higher scores. In contrast, in States with very low participation rates there is a greater tendency for outliers (relatively few significant results) to significantly distort the mean. In this particular situation, the mean for the low participation states is skewed positively, 
possibly due to the reasons mentioned above.
